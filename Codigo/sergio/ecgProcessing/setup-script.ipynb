```python
# Create test ECG data in array format 12 leads x 5000 samples
import numpy as np
ecg = np.random.rand(12, 5500)
sample_rate = 400
leads = ['I', 'II', 'III', 'aVR', 'aVL', 'aVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']
```

```python
from ecgprep import preprocess, read_ecg
import ecg_plot
ecg, sample_rate, leads = preprocess.preprocess_ecg(ecg, sample_rate, leads,
                                                    new_freq=500,
                                                    new_len=5000,
                                                    scale=1,
                                                    remove_powerline=True,
                                                    use_all_leads=True,
                                                    remove_baseline=True)

print(ecg.shape)

ecg_plot.plot(ecg, sample_rate=sample_rate,
                lead_index=leads, style='bw')
```
